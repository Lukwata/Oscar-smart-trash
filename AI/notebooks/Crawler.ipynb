```python
import os
from icrawler.builtin import GoogleImageCrawler
```

# Soft-cover books

```python
root_folder = "soft-cover_books"
keywords = ["soft cover book", "soft cover novel", "soft cover children books", "old soft-cover books"]
num_images = [300, 300, 100, 100]
```

```python
sum_images = 1
for keyword, max_num in zip(keywords, num_images):
    google_crawler = GoogleImageCrawler(parser_threads=4, downloader_threads=8,
                                        storage={'root_dir': os.path.join('./data/', root_folder, keyword.replace(" ", "_"))})
    
    google_crawler.crawl(keyword=keyword, max_num=max_num, offset=sum_images,
                         date_min=None, date_max=None,
                         min_size=(200,200), max_size=None)
    
    sum_images += max_num
```
