# Visualize some images

```python
candidate_df = desc_df[desc_df["desc"].str.lower().str.contains("folder".lower())]
candidate_df
```

```python
class_in_box = "/m/0663v"
```

```python
class_in_box = ""
for c in candidate_df.iterrows():
    index, series = c
    class_name, desc = series
    if class_name in box_classes:
        class_in_box = class_name
        print(class_name, desc)
```

```python
image_df = box_df[box_df["LabelName"] == class_in_box]
image_df
```

```python
root_path = "/media/ubuntu/Data/dataset/open-images/bbox/validation/"
image_ids = image_df["ImageID"].unique()
num_images = min(10, len(image_ids))
fig, axs = plt.subplots(num_images, figsize=(10, 6 * num_images))
for i, image_id in enumerate(image_ids[:num_images]):
    img = plt.imread(os.path.join(root_path, image_id + ".jpg"))
    axs[i].imshow(img)
```
